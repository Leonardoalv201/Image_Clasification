```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten , Conv2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import os
import random
import shutil
from sklearn.model_selection import train_test_split
```

```python
# Directorio original que contiene todas las imágenes clasificadas
original_dir = '/content/gdrive/MyDrive/IA 2.0/IA/Act 2.1/images'

# Directorio donde se crearán los subdirectorios de entrenamiento y prueba
base_dir_train = '/content/gdrive/MyDrive/IA 2.0/IA/Act 2.1/splitImg/train'

base_dir_test = '/content/gdrive/MyDrive/IA 2.0/IA/Act 2.1/splitImg/test'
```

```python
# Parámetros de preprocesamiento de imágenes
image_size = (150, 150)
batch_size = 8
```

###Aumento de datos

```python
# Aumento de datos
train_datagen = ImageDataGenerator(
    rescale = 1. / 255,
    rotation_range = 60,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    zoom_range = 0.2,  
    horizontal_flip = True  
)

# Aplicar aumento de datos en train
train_generator = train_datagen.flow_from_directory(
    base_dir_train,
    target_size = image_size,
    batch_size = batch_size,
    class_mode = 'binary'  
)
```

```python
test_datagen = ImageDataGenerator(rescale = 1. / 255)

test_generator = test_datagen.flow_from_directory(
    base_dir_test,
    target_size = image_size,
    batch_size = batch_size,
    class_mode = 'binary'  
)
```

###Print de imagenes modificadas

```python
import matplotlib.pyplot as plt

plt.figure()
#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(1, 5, figsize=(30, 8)) 

for i in range(5) :
  axarr[i].imshow(train_generator[0][0][0])
```

#Modelo base

Se usaron las siguientes capas:

*   Con2D
*   Flatten
*   Densa de 256 nodos y act relu
*   Densa de 1 nodo y act sigmord

Para compilar el modelo se usan como metricas:

*   Loss = binary_crossentropy
*   Optimizer = adam
*   Metrics = accuracy

```python
model1 = Sequential()
model1.add(Conv2D(10, (3, 3), activation = "relu", input_shape = (150,150,3)))
model1.add(Flatten())
model1.add(Dense(256, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))  

model1.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

##Train modelo base

```python
# Entrenar el modelo base
model1.fit(
    train_generator,
    epochs=10
)
```

##Test modelo base

```python
test_loss, test_acc = model1.evaluate(test_generator, steps = 25)
print('\ntest acc :\n', test_acc)
```

###Guardado del modelo

Se guarda el modelo para no tener que entrenarlo nuevamente

```python
model1.save('modeloBase.h5')
```

#Modelo mejorado

Para la mejora del modelo, se cambio la primer capa de un simple Conv2D a un modelo preentrenado VGG16, eliminando sus capas densas finales 

Se usaron las siguientes capas:

*   VGG16
*   Flatten
*   Densa de 256 nodos y act relu
*   Densa de 1 nodo y act sigmord

Para compilar el modelo se usan como metricas:

*   Loss = binary_crossentropy
*   Optimizer = adam
*   Metrics = accuracy

```python
# Cargar el modelo base VGG16 sin las capas densas finales y con pesos de ImageNet
base_model = VGG16(
    include_top = False,
    weights = 'imagenet',
    input_shape = (image_size[0], image_size[1], 3)
)

# Congelar los pesos de las capas base (no se entrenarán durante el ajuste fino)
base_model.trainable = False
 
model2 = Sequential()
model2.add(base_model)
model2.add(Flatten())
model2.add(Dense(256, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))  

model2.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

##Train modelo mejorado

```python
# Entrenar el modelo
model2.fit(
    train_generator,
    epochs=10
)
```

##Test modelo mejorado

```python
test_loss, test_acc = model2.evaluate(test_generator, steps = 25)
print('\ntest acc :\n', test_acc)
```

###Guardado del modelo

Se guarda el modelo para no tener que entrenarlo nuevamente

```python
model2.save('modeloFinal.h5')

#model = load_model('modeloFinal.h5')
```
